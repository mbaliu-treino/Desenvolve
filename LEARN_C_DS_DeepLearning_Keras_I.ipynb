Essas **funções** que irão extrair essas características são uma rede de unidades, uma rede neural. Elas que irão classificar as imagens.

> Enquanto o <font color=orange>modelo de Machine Learning</font> aprende com as características que nomeamos, as <font color=orange>funções de Deep Learning</font> *extraem* essas características, pois entendem quais características são essas. Lembrando que essa extração não precisa ser apenas para imagens ou problemas de classificação, podemos aplicar essa técnica para outros tipos de dados e problemas.

# REDES NEURAIS

Elas são extremamente poderosas por conseguirem resolver diversos tipos de problemas usando uma mesma ideia de processamento. Ela é usada para visão computacional, na identificação de objetos, identificar palavras e compreender textos, dados com a dimensão do tempo (séries temporais), edição de imagens. 

_Mas como ela consegue processar esses problemas?_
Um dos maiores desafios da computação é realizar uma <font color=orange><b>transformação nos dados</b></font>, de forma a possibilitar o seu processamento pelas máquinas de forma eficiente e acurada. Neste desafio, a busca de uma <font color=orange><b>função</b></font> que ajude na inferência é fundamental. Esta função transformará um dado em um outro dado/informação. 

É na obtenção dessa função que as redes neurais se desenpenham muito bem. Por isso elas são chamadas de aproximadores universais de funções, o qual se dá o nome de **Teorema da Aproximação Universal**. 

Nela é diga que uma rede neural com uma camada escondida é possível aproximar qualquer função contínua. Para conseguir se aproximar de qualquer função contínua, a rede neural se utiliza de inúmeras funções. Cada neurônio é a representação de uma função simples, não-linear ou linear.

Com duas funções não-lineares e com pesos invertidos é possível obter uma função torre/barra. No limite, um número infinito de torres e estreitas é possível construir uma função contínua extremamente complexa. Por isso é recomendado o uso de números pares de neurônios.

# Modelo

No processamento do Keras, os pixels da imagem serão convertidos em um *tensor*, uma estrutura linear achatada (1, n) - uma linha com n colunas. Isso será chamado de **camada 0**. Ao adicionar mais uma camada que terá comunicação com a camada anterior, criamos uma **camada densa**. 

> A camada dense (densa) é aquela que faz as conexões entre todas as unidades da camada 2 com as unidades da camada 1.

Cada conexão de funções, nódulo, será atribuído uma **probabilidade** para a classificação. Para isso, o algoritmo `softmax` irá obter a probabilidade para cada função e escolherá a de maior probabilidade.

- **Modelo Sequencia**
    - Entrada
    - Processamento
    - Saída

```python
# CONSTRUÇÃO DE MODELO DE REDES NEURAIS
modelo = keras.Sequential([
    # entrada - camada 0
    keras.layers.Flatten(input_shape = (28,28)),
    # processamento - camada 1
    keras.layers.Dense(256, activation=tensorflow.nn.relu),  # múltiplos de 2
    # saída - camada saída
    keras.layers.Dense(10, activation=tensorflow.nn.softmax)  # igual ao número de classes
])


modelo.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy')
# adam :: para variáveis multinomiais maiores do que 10
# sparse_categorical_crossentropy :: entropia categoria cruzada esparsa

modelo.fit(imagens_treino, identificadores_treino)
```

**ReLU - Unidade Linear Retificada**

A função ReLU introduz nas redes neurais modelos não-lineares, pois em seu gráfico não é possível representar com uma única linha. Os valores negativos são transformados em 0; os valores positivos se mantêm positivos. Na classificação com redes neurais ocorre um processo semelhante ao SVM, em que é buscado uma fórmula para criar áreas homogêneas.

Essas camadas intermediárias são chamadas também de camadas ocultas, pois não temos o conhecimento de como est[a sendo o processamento.

**Softmax**

Na última camada, o número de saídas deverá ser igual ao número de categorias que desejamos prever. O softmax irá calcular a probabilidade da imagem pertencer a cada uma delas .

Diminuir a abrangência numérica ajuda na qualidade do modelo.

```python
plt.imshow(imagens_treino[0], cmap='gray_r')
plt.colorbar()
```

```python
# NORMALIZAÇÃO DOS VALORES DOS PIXELS
# Normalização da entrada
imagens_treino_norm = imagens_treino/255.0

modelo = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(256, activation=tensorflow.nn.relu),  # múltiplos de 2
    keras.layers.Dense(10, activation=tensorflow.nn.softmax)  # igual ao número de classes
])

modelo.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy')

modelo.fit(imagens_treino_norm, identificadores_treino)
```

```python
# AUMENTANDO O NÚMERO DE CAMADAS
imagens_treino_norm = imagens_treino/255.0

modelo = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    # Número de camadas
    keras.layers.Dense(256, activation=tensorflow.nn.relu),  # múltiplos de 2
    keras.layers.Dense(128, activation=tensorflow.nn.relu),
    keras.layers.Dense(10, activation=tensorflow.nn.softmax)  # igual ao número de classes
])

modelo.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy')

modelo.fit(imagens_treino_norm, identificadores_treino)
```

```python
# AUMENTANDO O NÚMERO DE ÉPOCAS - repetições de treinamento
imagens_treino_norm = imagens_treino/255.0

modelo = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(256, activation=tensorflow.nn.relu),  # múltiplos de 2
    keras.layers.Dense(128, activation=tensorflow.nn.relu),
    keras.layers.Dense(10, activation=tensorflow.nn.softmax)  # igual ao número de classes
])

modelo.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy')

# Épocas
modelo.fit(imagens_treino_norm, identificadores_treino, epochs=5)
```

## Teste do Modelo


```python
# PREDIÇÃO
testes = modelo.predict(imagens_teste)
testes[0]
```

```python
# Posição do maior valor
id = 1

print('Resultado da predição de teste:', np.argmax(testes[id]))
print('Número da imagem de teste:', identificadores_teste[id])
```

## Avaliação

```python
# ACURÁCIA
imagens_treino_norm = imagens_treino/255.0

modelo = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(256, activation=tensorflow.nn.relu),  # múltiplos de 2
    keras.layers.Dense(128, activation=tensorflow.nn.relu),
    keras.layers.Dense(10, activation=tensorflow.nn.softmax)  # igual ao número de classes
])

modelo.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy',
    # Acurácia
    metrics=['accuracy'])

modelo.fit(imagens_treino_norm, identificadores_treino, epochs=5)
```

```python
# Perda e acurácia no TESTE
perda_teste, acuracia_teste = modelo.evaluate(imagens_teste, identificadores_teste)

print('Perda do teste:', round(perda_teste,4))
print(f'Acurácia do teste: {acuracia_teste:.4f}')
```

# Validação

**Overfitting**: Quando a performance do modelo para os dados de treino estão melhorando, mas os de treino apresenta uma piora ou uma estagnação, isso pode ser um indicativo de **overfitting**, em que o modelo está aprendendo muito as especificidades dos dados do treino e não está aprendendo a generalizar para os dados de teste.

> "[...] a curva de validação na qual o erro aumenta cada vez mais e na curva de treino que ele apenas reduz. Isso indica que nosso modelo está perdendo pouquíssima informação no treino e muita na validação. O que mostra que ele está servindo bem demais o modelo, overfitting."
