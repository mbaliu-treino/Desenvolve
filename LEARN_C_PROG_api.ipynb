# API

A API é um serviço web (webservice) para transicionar dados de forma mais eficiente e direta, facilitando a organização e a padronização da estrutura dos dados.

## API simples: CEP e de busca de endereço

<a href="https://www.youtube.com/watch?v=m72WIejruxI"><font color=gray size=2>Aula de referência</font></a>

Para obter o CEP de um endereço ou o contrário, obter o endereço de um CEP, temos esse webservice gratuíto.

* [API ViaCEP](https://viacep.com.br/)

Para acessar os dados, usamos um link, chamado de URI ou *endpoint*, que especifica as requisições webservice.

```python
# Construção do ENDPOINT
cep = '01001000'
endpoint = f'https://viacep.com.br/ws/{cep}/json/'
```

```python
# Validação do padrão para CEP
import re
cep = '04645015 '

# Remoção de pontuações
# cep.replace('-', '').replace(' ', '').replace('.', '')
regex_exp = re.compile(r'[0-9]')
cep = ''.join(re.findall(regex_exp, cep))

endpoint = f'https://viacep.com.br/ws/{cep}/json/'
```

```python
# Requisição de dados por API
import requests

if len(cep) == 8:
    req = requests.get(endpoint)
    dic_req = req.json()
    print(req)  # importante para verificar se está funcionando
    print(dic_req)
else:
    print('CEP inválido!')
```

```python
# Acessando aos dados do JSON
uf = dic_req['uf']
cidade = dic_req['localidade']
bairro = dic_req['bairro']

print(f'{bairro}, {cidade}, {uf}')
```

### Busca o CEP a partir do endereço

Buscará os resultados que mais se assemelham aos dados passados.

```python
uf = 'SP'
cidade = 'São Paulo'
logradouro = 'Antônio'

endpoint = f'https://viacep.com.br/ws/{uf}/{cidade}/{logradouro}/json/'


req = requests.get(endpoint)
dic_req = req.json()
print(req)  # importante para verificar se está funcionando
print(dic_req)
print(len(dic_req))
```
