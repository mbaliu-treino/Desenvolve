# Visualizando os dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
plt.style.use("seaborn")

mpl.rcParams['figure.figsize'] = (10,6)
mpl.rcParams['font.size'] = 12
```

```python

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris
```

```python
np.random.seed(123)

dados = iris.sample(50)
idx_teste = dados.sample(25).index
idx_treino = ~dados.index.isin(idx_teste)
```

```python
X = dados['petal_length']
y = dados['sepal_length']
```

```python
plt.scatter(X[idx_treino], y[idx_treino], color='coral')
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Comprimento da Sépala')
plt.show()
```

# Ajustado um modelo

O ajustamento excessivo no treino pode representar um caso de overfitting, pois o modelo se torna muito enviesado para o treino, não sabendo compreender o teste.

Nas árvores de decisão isso ocorre com o excesso de profundido, nos polinômios é o excesso de graus, nas redes neurais por serem consideradas Aproximadores Universais de Funções Contínuas, é o excesso de nodes/neurônios ou camadas.

Por tanto, muito parâmetros não representam uma boa qualidade do modelo, por sua baixa capacidade de generalização, performando muit mal na previsão de amostras futuras, que ainda não foram vistas.

Por tanto, o deve-se ter cuidado com underfitting e o overfitting, mas fazendo valer o princípio da Navalha de Ockham e sempre buscando o modelo mais simples.

```python
grau = 0
z = np.polyfit(X[idx_treino], y[idx_treino], grau)
p = np.poly1d(z)

plt.scatter(X[idx_treino], y[idx_treino], color='coral')
sns.lineplot(x=X[idx_treino], y=p(X[idx_treino]), color='k')
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Comprimento da Sépala')
plt.show()
```

# Validação do modelo

```python
grau = 300
z = np.polyfit(X[idx_treino], y[idx_treino], grau)
p = np.poly1d(z)

plt.scatter(X[idx_treino], y[idx_treino], color='coral', label='Treino')
sns.lineplot(x=X[idx_treino], y=p(X[idx_treino]), color='k')
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Comprimento da Sépala')

plt.scatter(X[idx_teste], y[idx_teste], color='C', label='Teste')
plt.legend(fontsize=16)
plt.show()
```
