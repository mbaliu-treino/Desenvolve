## 4 - Análise em Streaming

Com os dados do Twitter sendo obtidos em Streaming, podemos também analisá-los na Azure também em Streaming. Isso permite que acompanhêmos um determinado tema. Quando algo for encontrado, podemos realizar alguma ação, como mandar um e-mail, um SMS, gravar uma informação em log ou executar outro código.

```python
from azure.cognitiveservices.vision.computervision.models import VisualFeatureTypes
import json

# herança da biblioteca StreamListener
class MinhaStreamListener(tw.StreamingClient):
    def on_tweet(self, tweet):
        print('Usuário: ', tweet.user.screen_user)
        print('Texto: ', tweet.text)
        # print(tweet.attachments)
        # print(tweet.entities)
        if 'media' in tweet.entities:
            url = tweet.entities['media'][0]['media_url']
            print('URL: ', url)

            analise_marcos = client_cv.analyze_image(url, VisualFeatureTypes, details=['Landmarks'])
            analise_marcos = analise_marcos.as_dict()
            nome_marco = analise_marcos['categories'][0]['detail']['landmarks'][0]['name']
            print('Ponto turistico: ', nome_marco)

            descricao = client_cv.describe_image(url, 1, 'en')
            # texto_descricao = 
            print('Descrição: ', descricao)

            lista_tags = [tag['name'] for tag in analise_marcos['tags']]
            print('Tags: ', lista_tags)

            resultados = {
                'usuario':tweet.user.screen_user,
                'texto': tweet.text,
                'url': url,
                'marco': nome_marco,
                'descricao': descricao,
                'lista_tags': lista_tags
            }

            # Salvando dicionário
            with open('tweets.txt', 'a') as arquivo:
                arquivo.write(json.dumps(resultados))
                arquivo.write('\n')


        print('--------')
        print('\n')

lugares = ['landmark', '#photo', '#landscape']
minhasStream = MinhaStreamListener(bearer_token=bearer_token)
# O limite de regras é 25
for l in lugares:
    minhasStream.add_rules(tw.StreamRule(l))

# minhasStream.filter()
len(minhasStream.get_rules().data)
```

## 4 - Análise de Tweets (Twitter + Azure Computer Vision)

A análise feita foi identificação de imagens que abordam um tema determinado, possibilitando selecionar um conjunto de imagens que é o do nosso interesse. Para isso foi realizada a coleta de imagens do Twitter e a classificação de tag usando a funcionalidade de visão computacional da Microsoft Azure.

```python
import json
from azure.cognitiveservices.vision.computervision.models import VisualFeatureTypes

cidades = ['berlin', 'barcelona', 'paris', 'lisboa', 'munich', 'hamburg', 'madrid']

# Realização de pesquisa
for cidade in cidades:
    tweets = tw.Cursor(
        api.search_tweets,
        q=cidade,).items(20)
    # print('Cidade: ', cidade)

    # print('Imagens postadas:')
    for i, tweet in enumerate(tweets):
        # print(i)
        if 'media' in tweet.entities:
            # print('Usuário: ', tweet.user.screen_name)  # quem postou a imagem
            url = tweet.entities['media'][0]['media_url']
            # print('Imagem: ', url)  # url da mídia
            # print('Texto: ', tweet.text) 

            descricao = client_cv.describe_image(url, 1, 'en')
            descricao = descricao.captions[0].text
            # print('Descrição: ', descricao)

            landmarks_analysis = client_cv.analyze_image(url, VisualFeatureTypes, details=['Landmarks']).as_dict()
            for categoria in landmarks_analysis['categories']:
                marco_value = 0
                marco = ''
                if 'name' in categoria:
                    if categoria['score'] > marco_value:
                        marco = categoria['name']
                        marco_value = categoria['score']
                    if 'detail' in categoria:
                        if categoria['detail']['landmarks']:
                            marco = categoria['detail']['landmarks'][0]['name']
            # print('Ponto turístico: ', marco)

            tags = [tag['name'] for tag in landmarks_analysis['tags']]
            # print('Tags: ', tags)

            resultado = {
                'cidade': cidade,
                'user': tweet.user.screen_name,
                'imagem': tweet.entities['media'][0]['media_url'],
                'texto': tweet.text,
                'tags': tags,
                'descricao': descricao,
                'marco_turistico': marco
            }
            print(resultado)

            # Salvando dicionário
            with open('tweets.txt', 'a') as arquivo:
                arquivo.write(json.dumps(resultado))
                arquivo.write('\n')

    # print('-------')
    # print('\n')

            
```
